```python
import matplotlib.pyplot as plt  
import numpy as np
import sympy as sy
from sympy import Eq, solve, Symbol, symbols, simplify
from sympy.plotting import plot
%matplotlib inline
```

```python
var_cv1 = 10 #require 基礎消費
var_cv2 = 0.6 #require 限界消費性向
var_iv1 = 10 #require 投資係数
var_G = 100 #require 政府支出
var_YF = 200 #require 完全雇用国民所得

C=Symbol('C') #require
I=Symbol('I') #require
_cv1=Symbol('_cv1') #require
_cv2=Symbol('_cv2') #require
_iv1=Symbol('_iv1') #require
G=Symbol('G') #require
YF=Symbol('YF') #require
G_delta = Symbol('G_delta') #require

#IS_line
C = _cv1 + (_cv2*YF)
I = _iv1
Y_eq = C + I + G + G_delta - YF
Y_eq = simplify(Y_eq)
print("eq_yield is",Y_eq)
print()
sub_eq = Y_eq.subs([(_cv1,var_cv1),(_cv2,var_cv2),(_iv1,var_iv1),(G,var_G),(YF,var_YF)])
print("sub_eq is",sub_eq)
answer = solve(sub_eq, G_delta)
print()
print("answer:G_delta is",float(answer[0]))
print()
if float(answer[0]) < 0:
    print("財務大臣のアドバイス：完全雇用国民所得が{}に対し、現状の財政支出は{}単位ですが、政府支出が多すぎるため財政支出を減らす必要があります目安は{}単位だけ財政支出を増減させてください".format(var_YF, var_G, float(answer[0])))
if float(answer[0]) == 0:
    print("財務大臣のアドバイス：完全雇用国民所得が{}に対し、現状の財政支出は{}単位となっており、これは政府支出はちょうど適正ラインです".format(var_YF, var_G))

if float(answer[0]) > 0:
    print("財務大臣のアドバイス：完全雇用国民所得が{}に対し、現状の財政支出は{}単位ですが、政府支出が少なすぎるため、さらなる財政支出が必要です。目安は{}単位だけ財政支出を増減させてください".format(var_YF, var_G, float(answer[0])))

p1 = plot(sub_eq, (G_delta, -2*(float(answer[0])), 2*(float(answer[0]))))
p1
```
